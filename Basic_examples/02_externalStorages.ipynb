# External Storages
External Storages can be used for heat transer over the boundaries of the system. For external storages the temperature stays as set.

```python
# model setup
CP_WATER = 4182 # J/(kg*K)
from ThermoBuilPy import *
cap1 = ThermalStorage.newStorage(cap=1000,temp=30,name='Capacity 1')

extStor1 = ExtStorage.newExtStorage(temp=50) # <------- Seting up an external storage

cond1 = Conduction.newConduction(cap1,extStor1,100)
ts = ThermalSystem.newThermalSystem(storages=[cap1],
                                    extStorages=[extStor1],
                                    conductions=[cond1])
ts.simulate(100,1)
```

#### Changing the temperature of an external storage

```python
extStor1.set_temp(20)
```

```python
ts.simulate(100,1)
```

```python
ts.plot_temps([cap1])
```
