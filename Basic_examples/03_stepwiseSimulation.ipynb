# Stepwise Simulation
In the 01_basicExcample the method ThermalSystem.simulate() is used for the the simulation of the thermal RC-system for a time series with given stepsize and number of steps. In this case, the package is working with np.arrays for storing the time series results.

However, it is also possible to run the simulation step by step (or just for one time step). In this case, as the total number of stepsis not set from the beginning, the results are stored in lists. The simulation must be prepared using the prepare_simulation-method. After that single steps can be calculated by the use of the do_simstep-method.

```python
# model setup
CP_WATER = 4182 # J/(kg*K)
from ThermoBuilPy import *
cap1 = ThermalStorage.newStorage(cap=1000,temp=30,name='Capacity 1')
extStor1 = ExtStorage.newExtStorage(temp=50)
cond1 = Conduction.newConduction(cap1,extStor1,100)
ts = ThermalSystem.newThermalSystem(storages=[cap1],
                                    extStorages=[extStor1],
                                    conductions=[cond1])
```

#### Preparing the simulation

```python
ts.prepare_simulation(stepsize=1)
```

#### Running time steps

```python
for t in range(100):
    ts.do_simstep()

extStor1.set_temp(20)
for t in range(100):
    ts.do_simstep()
```
